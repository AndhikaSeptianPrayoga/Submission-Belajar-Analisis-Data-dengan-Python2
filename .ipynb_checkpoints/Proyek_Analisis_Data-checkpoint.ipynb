### Cleaning Data

```python
# Konversi suhu dari skala normalisasi ke Celsius
day_df['temp_celsius'] = day_df['temp'] * 41
day_df['atemp_celsius'] = day_df['atemp'] * 50

hour_df['temp_celsius'] = hour_df['temp'] * 41
hour_df['atemp_celsius'] = hour_df['atemp'] * 50
```

```python
# Menghitung median kelembapan
median_hum = day_df[day_df['hum'] > 0]['hum'].median()

# Mengganti nilai 0 dengan median
day_df['hum'] = day_df['hum'].replace(0, median_hum)
hour_df['hum'] = hour_df['hum'].replace(0, median_hum)
```

```python
# Menentukan batas outlier menggunakan IQR (Interquartile Range)
Q1 = day_df['cnt'].quantile(0.25)
Q3 = day_df['cnt'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Menampilkan jumlah outlier
outliers = day_df[(day_df['cnt'] < lower_bound) | (day_df['cnt'] > upper_bound)]
print(f"Jumlah outlier dalam 'cnt': {outliers.shape[0]}")

# Opsi: Menghapus outlier (Jika perlu)
day_df_cleaned = day_df[(day_df['cnt'] >= lower_bound) & (day_df['cnt'] <= upper_bound)]
```

**Insight:**
- Suhu diubah ke dalam Celsius, tidak ada kelembapan 0%, dan outlier sudah diperiksa.

Selain cuacanya lebih hangat, kita bisa menghubungkan juga dengan banyaknya hari libur atau banyaknya cuti yang terjadi di pertengahan tahun atau musim panas. Namun terdapat outlier dalam boxplot di musim dingin, dan musim gugur, mari kita analisis lebih lanjut lagi menggunakan stripplot.

```python
fig, ax = plt.subplots(figsize=(8, 5))

sns.boxplot(x=day_df["season_label"], y=day_df["cnt"], palette="Set2", ax=ax)
sns.stripplot(x=day_df["season_label"], y=day_df["cnt"], color="black", jitter=True, alpha=0.4, ax=ax)

ax.set_title("Distribusi Penyewaan Sepeda Berdasarkan Musim")
ax.set_xlabel("Musim")
ax.set_ylabel("Jumlah Penyewaan Sepeda")

plt.show()
```

**Insight:**
- Setelah kita lihat kembali hanya terdapat sedikit outlier di dalam boxplot musim dingin dan musim gugur. Walaupun berjumlah sangat sedikit, mari kita hilangkan nilai outlier tersebut agar tidak mengganggu hasil akhir dari analisis data ini.

```python
# Fungsi untuk menghapus outlier per musim menggunakan IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Cek jumlah data sebelum penghapusan outlier
print(f"Jumlah data sebelum pembersihan: {len(day_df)}")

# Menghapus outlier per musim secara permanen
day_df_clean = day_df.groupby("season_label", group_keys=False).apply(remove_outliers_iqr, column="cnt")

# Cek jumlah data setelah penghapusan outlier
print(f"Jumlah data setelah pembersihan: {len(day_df_clean)}")

# Plot boxplot dengan stripplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=day_df_clean["season_label"], y=day_df_clean["cnt"], palette="Set2")

# Menambahkan stripplot agar titik data individual terlihat
sns.stripplot(x=day_df_clean["season_label"], y=day_df_clean["cnt"], color="black", alpha=0.4, jitter=True)

plt.title("Distribusi Penyewaan Sepeda Berdasarkan Musim (Tanpa Outlier)")
plt.xlabel("Musim")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.show()
```

**Insight:**
- Kita telah berhasil mereduksi nilai outlier pada boxplot musim dingin dan musim gugur. Walaupun masih ada sedikit nilai outlier, tetapi kita tidak dapat menggunakan quantile lebih rendah lagi sebab akan membuat data menjadi underfitting.

**Insight:**
- Dari dua plot di atas, kita dapat mengambil insight bahwa kebanyakan penyewa menyewa sepeda pada jam-jam produktif. Yaitu pada pagi hari ketika berangkat kerja dan pada sore hari ketika pulang kerja. Terdapat sedikit sekali penyewa, yang menyewa sepeda pada jam-jam non-produktif seperti tengah malam hingga dini hari. Namun masih terdapat banyak nilai outlier yang ada pada data ini, mari kita reduksi.

```python
# Fungsi untuk menghapus outlier menggunakan IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Menghapus outlier per jam
hour_df_clean = hour_df.groupby("hr", group_keys=False).apply(remove_outliers_iqr, column="cnt")

# Membuat subplot dengan 1 lineplot dan 1 boxplot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Lineplot setelah menghapus outlier
sns.lineplot(x=hour_df_clean['hr'], y=hour_df_clean['cnt'], marker='o', color='b', ax=axes[0])
axes[0].set_title("Tren Penyewaan Sepeda Berdasarkan Jam dalam Sehari")
axes[0].set_xlabel("Jam dalam Sehari (0-23)")
axes[0].set_ylabel("Jumlah Penyewaan Sepeda")
axes[0].set_xticks(range(0, 24))
axes[0].grid(True)

# Boxplot setelah menghapus outlier
sns.boxplot(x=hour_df_clean["hr"], y=hour_df_clean["cnt"], palette="Set2", ax=axes[1])
axes[1].set_title("Distribusi Penyewaan Sepeda Berdasarkan Jam")
axes[1].set_xlabel("Jam dalam Sehari (0-23)")
axes[1].set_ylabel("Jumlah Penyewaan Sepeda")
axes[1].set_xticks(range(0, 24))

# Menyesuaikan layout agar tidak bertumpuk
plt.tight_layout()
plt.show()
```

**Insight:**
- Setelah fase reduksi nilai outlier selesai, kita dapat melihat plot boxplot di kanan menunjukkan pengurangan drastis nilai outlier dari data distribusi penyewaan sepeda berdasarkan jam, dengan quantile q1 = 25% dan q3 75%. Kami tidak dapat menurunkan quantile lebih rendah, karena akan beresiko menjadikan data underfitting, dan dapat menghilangkan data yang sebenarnya valid.

### **Faktor Apa yang Paling Berpengaruh terhadap Penyewaan Sepeda?**

```python
numeric_df = day_df.select_dtypes(include=['number'])  # Hanya kolom numerik
correlation = numeric_df.corr()

print("\nKorelasi Faktor terhadap Penyewaan Sepeda:\n", correlation)

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Korelasi Antar Faktor dalam Penyewaan Sepeda")
plt.show()
```

**Insight:**

Lalu, Temperature (0.66) dan atemp (0.66) yang berarti semakin tinggi suhu, semakin banyak sepeda yang disewa. Ini menunjukkan bahwa cuaca hangat lebih mendukung aktivitas bersepeda. Season (0.41) yang berarti musim cukup memengaruhi jumlah penyewaan, kemungkinan dari EDA sebelumnya, bahwa banyak penyewa menyewa sepeda di musim semi dan panas.

### **Perbandingan Jumlah Pengguna Registered dan Casual**

```python
# Mapping hari dalam seminggu
dayofweek_map = {0: "Sen", 1: "Sel", 2: "Rab", 3: "Kam", 4: "Jum", 5: "Sab", 6: "Min"}
day_df["weekday_label"] = day_df["weekday"].map(dayofweek_map)

# Urutan hari agar sesuai Senin-Minggu
weekday_order = ["Sen", "Sel", "Rab", "Kam", "Jum", "Sab", "Min"]

# Lineplot: Tren Rata-rata Penyewaan Registered vs Casual
weekday_agg = day_df.groupby("weekday_label", observed=True)[["registered", "casual"]].mean().reindex(weekday_order)

plt.figure(figsize=(8, 5))
sns.lineplot(x=weekday_agg.index, y=weekday_agg["registered"], marker="o", color="blue", label="Registered")
sns.lineplot(x=weekday_agg.index, y=weekday_agg["casual"], marker="s", color="red", label="Casual")
plt.title("Tren Penyewaan Registered vs Casual Berdasarkan Hari")
plt.xlabel("Hari dalam Seminggu")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.legend()
plt.tight_layout()
plt.show()
```

**Insight:**
- Grafik di atas menunjukkan bahwa user yang tidak terdaftar sering menggunakan sepeda pada weekend, hal ini dapat dipahami karena jenis pengguna ini kemungkinan memang hanya membutuhkan sepeda untuk menikmati hari liburnya. Berbeda dengan pengguna registered, mereka hampir menyewa sepeda setiap hari. Oleh karena itu, dibutuhkan untuk mendaftar dengan kemungkinan mendapat fee sewa yang lebih murah dibanding pengguna casual.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh cuaca terhadap jumlah penyewaan sepeda?


```python
# Mapping kategori cuaca
weather_labels = {1: "Cerah", 2: "Mendung", 3: "Hujan Ringan", 4: "Hujan Lebat"}
day_df["weather_label"] = day_df["weathersit"].map(weather_labels)

# Menghitung rata-rata penyewaan sepeda berdasarkan kondisi cuaca
weather_agg = day_df.groupby("weather_label")["cnt"].mean().reset_index()

# Membuat visualisasi dengan barplot
plt.figure(figsize=(10, 6))
sns.barplot(data=weather_agg, x="weather_label", y="cnt", palette="coolwarm")
plt.title("Pengaruh Cuaca terhadap Jumlah Penyewaan Sepeda", fontsize=14)
plt.xlabel("Kondisi Cuaca", fontsize=12)
plt.ylabel("Rata-rata Penyewaan Sepeda", fontsize=12)
plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Pada cuaca cerah, jumlah rata-rata penyewaan sepeda memiliki rata-rata paling tinggi, berbeda dengan cuaca Hujan Ringan yang memiliki rata-rata penyewaan yang paling rendah. Hal ini tentu berkorelasi dengan enggannya para penyewa untuk _basah-basahan_ memakai sepeda ketika cuaca sedang hujan.

### Pertanyaan 2: Apakah ada tren musiman dalam penggunaan sepeda (musim semi, panas, gugur, dan dingin)?

```python
# Mapping musim
season_map = {1: "Musim Semi", 2: "Musim Panas", 3: "Musim Gugur", 4: "Musim Dingin"}
day_df["season_label"] = day_df["season"].map(season_map)

# Rata-rata jumlah penyewaan sepeda per musim
season_agg = day_df.groupby("season_label", observed=True)["cnt"].mean().reset_index()

# Membuat subplot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Barplot - Rata-rata penyewaan per musim
sns.barplot(data=season_agg, x="season_label", y="cnt", palette="coolwarm", ax=axes[0])
axes[0].set_title("Rata-rata Penyewaan Sepeda Berdasarkan Musim")
axes[0].set_xlabel("Musim")
axes[0].set_ylabel("Jumlah Penyewaan Sepeda")

# Lineplot - Tren penyewaan sepanjang musim
sns.lineplot(data=day_df, x="season_label", y="cnt", estimator="mean", ci=None, marker="o", palette="coolwarm", ax=axes[1])
axes[1].set_title("Tren Penyewaan Sepeda Sepanjang Musim")
axes[1].set_xlabel("Musim")
axes[1].set_ylabel("Jumlah Penyewaan Sepeda")

plt.tight_layout()
plt.show()
```

**Insight:**
- Musim gugur memiliki rata-rata penyewaan tertinggi, sedikit lebih tinggi dari musim panas. Ini menunjukkan bahwa pada hari-hari tertentu di musim gugur, aktivitas bersepeda cukup tinggi. Musim dingin juga memiliki rata-rata penyewaan yang cukup tinggi, menandakan bahwa meskipun cuaca dingin, masih ada banyak pengguna setia yang menyewa sepeda. Musim semi memiliki rata-rata penyewaan terendah, kemungkinan karena cuaca yang masih tidak menentu. Ini sangat berbeda apabila dibandingkan dengan rata-rata penyewa di musim yang berbeda, di mana rata-rata jumlah penyewa berada di musim panas dan terendah di musim dingin, hal ini dapat terjadi karena rentang waktu per musim yang berbeda-beda, dan adanya variasi dalam data yang memengaruhi perbedaan antara jumlah dan rata-rata penyewa.

### Pertanyaan 3: Bagaimana pola penggunaan sepeda di hari kerja dibandingkan dengan akhir pekan?

```python
# Menambahkan kolom kategori: Hari Kerja vs Akhir Pekan
day_df["is_weekend"] = day_df["weekday"].apply(lambda x: "Akhir Pekan" if x in [5, 6] else "Hari Kerja")

# Menghitung rata-rata penyewaan untuk hari kerja vs akhir pekan
workday_agg = day_df.groupby("is_weekend")[["cnt", "registered", "casual"]].mean()

# Membuat Barplot untuk perbandingan penyewaan sepeda
plt.figure(figsize=(8, 5))
workday_agg.plot(kind="bar", color=["royalblue", "deepskyblue", "tomato"], alpha=0.8)
plt.title("Perbandingan Penyewaan Sepeda di Hari Kerja vs Akhir Pekan")
plt.xlabel("")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.xticks(rotation=0)
plt.legend(["Total Penyewaan", "Registered", "Casual"])
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.show()
```

**Insight:**

### Pertanyaan 4: Bagaimana tren penggunaan sepeda berdasarkan jam dalam sehari dan seminggu?

### Pertanyaan 5: Faktor apa yang paling berpengaruh terhadap jumlah penyewaan sepeda?

```python
# Memilih fitur yang diduga memiliki pengaruh terhadap jumlah penyewaan sepeda
correlation_features = ["temp", "hum", "windspeed", "registered", "casual"]
correlation_matrix = day_df[correlation_features + ["cnt"]].corr()

# Mengurutkan faktor berdasarkan korelasi terhadap 'cnt' (jumlah penyewaan)
correlation_sorted = correlation_matrix["cnt"].drop("cnt").sort_values(ascending=False)

# Membuat barplot untuk menunjukkan faktor paling berpengaruh
plt.figure(figsize=(8, 5))
sns.barplot(x=correlation_sorted.values, y=correlation_sorted.index, palette="coolwarm", orient="h")
plt.title("Faktor yang Paling Berpengaruh terhadap Penyewaan Sepeda")
plt.xlabel("Korelasi dengan Jumlah Penyewaan Sepeda")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

**Insight:**
- Faktor yang paling berpengaruh terhadap jumlah penyewaan sepeda adalah jenis pengguna (registered dan casual) dan suhu (temp) karena memiliki korelasi paling tinggi.

## Analisis Lanjutan (Opsional)

1. Time Series Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Dekompisi tren, musiman, dan residual
decomposition = seasonal_decompose(day_df["cnt"], model="additive", period=30)

# Plot hasil dekomposisi
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(decomposition.observed, label="Observed")
plt.legend()

plt.subplot(412)
plt.plot(decomposition.trend, label="Trend")
plt.legend()

plt.subplot(413)
plt.plot(decomposition.seasonal, label="Seasonal")
plt.legend()

plt.subplot(414)
plt.plot(decomposition.resid, label="Residual")
plt.legend()

plt.tight_layout()
plt.show()
```

**Insight:**
- Penyewaan sepeda cenderung meningkat seiring waktu, tetapi mengalami sedikit penurunan di akhir. Ada pola berulang yang menunjukkan kemungkinan pengaruh hari dalam seminggu atau musim tertentu terhadap penyewaan. Ada fluktuasi yang tidak dapat dijelaskan oleh tren dan pola musiman, yang bisa berasal dari faktor eksternal seperti acara khusus atau kondisi ekonomi.
